```python
import numpy as np
import matplotlib.pyplot as plt

from pd4ml import Spinodal

print('imports dones')
```

```python
# dataset description
Spinodal.print_description()
```

```python
# load training and testing set
X_train, y_train = Spinodal.load('train', path = '../datasets')
X_test, y_test = Spinodal.load('test', path = '../datasets')
```

```python
# checking basic content

print(len(X_train))
print(len(X_test))

print(X_train[0].shape)

print(y_train.shape)
print(y_test.shape)
```

# some basic plots

```python
# example 2d histograms

X = X_train[0]
y = y_train

fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(20, 15))
for i, (ax) in enumerate(np.array(axs).ravel()):
    ax.imshow(X[i,:,:])
    ax.set_title('2d hist No. {} - label = {}'.format(i, y[i]))
```
